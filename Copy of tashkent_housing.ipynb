```python
import pandas as pd
import numpy as np
import joblib
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/housing_tashkent.csv", index_col=0)
df
```

```python
from sklearn.model_selection import train_test_split
train_set ,test_set = train_test_split(df, test_size=0.2, random_state=42)
train_set.head(10)
```

```python
x_values = train_set.copy()
x_values.head()
```

```python
y = x_values['price']
y
```

```python
x_values.drop(columns='price', axis=1, inplace=True)
x_values.head()
```

```python
from sklearn.preprocessing import StandardScaler as sscaler
stand_scaler = sscaler()
```

```python
x_scaler = stand_scaler.fit_transform(x_values)
x_scaler
```

#LinearRegression

```python
from sklearn.linear_model import LinearRegression
LR_model = LinearRegression()
LR_model.fit(x_scaler, y)

filename = 'LR_model.jbl'
joblib.dump(LR_model, filename)
```

#DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor
Tree_model = DecisionTreeRegressor()
Tree_model.fit(x_scaler, y)

filename = 'Tree_model.jbl'
joblib.dump(Tree_model, filename)
```

#RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
RF_model = RandomForestRegressor()
RF_model.fit(x_scaler, y)

filename = 'RF_model.jbl'
joblib.dump(RF_model, filename)
```

```python
test_set
x_test = test_set.drop(columns='price', axis=1)
x_test
```

```python
y_test = test_set['price'].copy()
y_test
```

```python
x_test_prep = stand_scaler.fit_transform(x_test)
x_test_prep
```

```python
ln_prepared = LR_model.predict(x_test_prep)
tr_prepared=Tree_model.predict(x_test_prep)
rf_prepared = RF_model.predict(x_test_prep)
```

```python
d = {"LinearRegression": ln_prepared,"DecisionTreeRegressor": tr_prepared, "RandomForestRegressor": rf_prepared, "реальная цена": y_test}
pd.DataFrame(data=d)
```

```python
from sklearn.metrics import mean_squared_error
slr_error = mean_squared_error(y_test, ln_prepared)
str_error = mean_squared_error(y_test, tr_prepared)
srf_error = mean_squared_error(y_test, rf_prepared)
print(f"средняя квадратичная ошибка(LinearRegression) {np.sqrt(slr_error)}")
print(f"средняя квадратичная ошибка(DecisionTreeRegressor) {np.sqrt(str_error)}")
print(f"средняя квадратичная ошибка(RandomForestRegressor) {np.sqrt(srf_error)}")
```

```python

from sklearn.metrics import mean_absolute_error
lr_error = mean_absolute_error(y_test, ln_prepared)
tr_error = mean_absolute_error(y_test, tr_prepared)
rf_error = mean_absolute_error(y_test, rf_prepared)
print(f"средняя абсолютная ошибка(LinearRegression) {lr_error}")
print(f"средняя абсолютная ошибка(DecisionTreeRegressor) {tr_error}")
print(f"средняя абсолютная ошибка(RandomForestRegressor) {rf_error}")
```
